```
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```
df5044 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasetfb.xlsx')
```

```
df5044
```

```
x = df5044['Goals (X)']
y = df5044['Wins (Y)']
```

```
regr = linear_model.LinearRegression()
regr.fit(x.values.reshape(-1,1), y)

print('Nilai a (intercept) =', regr.intercept_)
print('Nilai b (coefficients) =', regr.coef_)
```

```
new_durasi = 30
print('Hasil prediksi kemenangan dengan gol 30 = \n', regr.predict([[new_durasi]]))
```

```
#Membuat Data Frame
df5044 = pd.DataFrame(list(zip(x,y)), columns = ['Goals (X)', 'Wins (Y)'])
df5044
```

```
df5044.describe()
```

```
df5044.mean()
```

```
mean5044 = np.mean(x)
mean5044
```

```
x_mean5044 = df5044['Goals (X)'] - df5044['Goals (X)'].mean()
print(x_mean5044)
y_mean5044 = df5044['Wins (Y)'] - df5044['Wins (Y)'].mean()
print(y_mean5044)
```

```
sum_5044 = np.sum(x_mean5044 * y_mean5044)
sum_5044
```

```
sqrt_5044 = np.sqrt(np.sum(x_mean5044**2) * np.sum(y_mean5044**2))
sqrt_5044
```

```
correlation5044 = sum_5044 / sqrt_5044
correlation5044
```

```
plt.title('Korelasi Variabel X dan Y')
plt.scatter(df5044['Goals (X)'] , df5044['Wins (Y)'])
plt.xlabel('Goals (X)')
plt.ylabel('Wins (Y)')
plt.show()
```

```
np.corrcoef(df5044['Goals (X)'] , df5044['Wins (Y)'])
```
